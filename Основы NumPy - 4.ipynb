```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.random.beta(0.1, 0.6, size=100)    #  бета распределение
a.sort()
plt.plot(a)
```

```python
a = np.random.gamma(shape=0.8, scale=1.7, size=100)    # гамма распределение
a.sort()
plt.plot(a)
```

```python
a = np.random.pareto(3.5, size=100)    #  Паретто распределение
a.sort()
plt.plot(a)
```

```python
a = np.random.chisquare(2.2, size=100)     #  хи-квадрат распределение
a.sort()
plt.plot(a)
```
